```python
# @title
import os
from IPython.core.magic import register_cell_magic

@register_cell_magic #(%%backend) oluştur sağlar.
def backend(line, cell):#Bu fonksiyon iki parametre alır:line: Hücrenin başına yazdığınız argüman (örneğin: %%backend jax yazarsanız, line değeri "jax" olur).

                        #cell: O hücrenin içindeki tüm kodlar.
    current, required = os.environ.get("KERAS_BACKEND", ""), line.split()[-1]
    if current == required:
        get_ipython().run_cell(cell)
    else:
        print(
            f"This cell requires the {required} backend. To run it, change KERAS_BACKEND to "
            f"\"{required}\" at the top of the notebook, restart the runtime, and rerun the notebook."
        )
```

```python
#Image classification
#Introduction to ConvNets


import keras
from keras import layers

inputs = keras.Input(shape=(28, 28, 1)) #Anlamı: Modelin $28 \times 28$ piksel boyutunda ve tek kanallı (siyah-beyaz) görselleri kabul edeceğini söyler. Eğer görseller renkli olsaydı, son rakam 3 (RGB) olurdu.
x = layers.Conv2D(filters=64, kernel_size=3, activation="relu")(inputs)#Filtreler (64, 128, 256): Ağ ilerledikçe filtre sayısı artar
x = layers.MaxPooling2D(pool_size=2)(x) #Kernel Size (3): Görüntü üzerinde gezen $3 \times 3$ boyutundaki küçük pencerelerdir.
x = layers.Conv2D(filters=128, kernel_size=3, activation="relu")(x)#ReLU: Negatif değerleri sıfırlayarak modelin doğrusal olmayan karmaşık ilişkileri öğrenmesini sağlar.
x = layers.MaxPooling2D(pool_size=2)(x)
x = layers.Conv2D(filters=256, kernel_size=3, activation="relu")(x)
x = layers.GlobalAveragePooling2D()(x) #Modern Yaklaşım: Eskiden burada tüm veriyi tek bir sütuna dizen Flatten kullanılırdı. GlobalAveragePooling ise her filtrenin ortalamasını alarak veriyi küçültür. Bu, modelin parametre sayısını azaltır ve aşırı öğrenmeyi (overfitting) engeller.
outputs = layers.Dense(10, activation="softmax")(x) #Softmax: 10 farklı sınıf (örneğin 0'dan 9'a kadar rakamlar) için olasılık üretir. Tüm çıktıların toplamı 1 olur; en yüksek olasılığa sahip sınıf, modelin tahminidir.
model = keras.Model(inputs=inputs, outputs=outputs)
```
